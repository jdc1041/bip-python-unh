```python
import pandas 
```

```python
from glob import glob
```

```python
f1 = pandas.read_excel("customers.xlsx")
f2 = pandas.read_excel("contracts.xlsx")
```

```python
result= pandas.merge(f2,f1,on = "customer_id",how='inner',indicator=True)
result.head()
```

```python
result.groupby("state")["monthly_amt", "deposit"].sum()
```

```python
result['Total'] = result.monthly_amt * result.contract_term
```

```python
result['total'] = result['Total'] + result['deposit']
result.groupby("state")["total"].sum()
```
